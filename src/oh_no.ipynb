```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
frames = []
if not cap.isOpened():
    print("Error: Could not open video file.")
while True:
    ret, frame = cap.read() # Reads a frame
    frames.append(frame)
    if not ret:
        break
    cv2.imshow("Video Frame", frame)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break


        
cv2.destroyAllWindows()
cap.release()
```
